```python
def min_num(num_list):
    result = num_list[0]
    for i in num_list:
        if result > i:
            result = i
    return result
min_num([52,273,32,103,275])    
```

```python
# 윤년을 확인하는 함수 만들기
# 4로 나누어 떨어지는 해는 윤년
# but 100으로 나누어 떨어지는 해는 윤년이 아님
# but 400으로 나누어 떨어지는 해는 윤년

def isLeapYear(num):
    if num % 4 == 0 and num % 100 != 100 or num % 400 == 0 :
        result = True
    else :
        result = False
    return result

isLeapYear(2024)
```
