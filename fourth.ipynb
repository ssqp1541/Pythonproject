소수판별함수  is_prime()
[2,3,4,5,6,7,8,9] -> [True, True, False, True, False, True, False, False]

```python
def is_prime(num_list):
    result = []
    for num in num_list:
        isprime = True
        for i in range(2,num):
            if num % i == 0:
                isprime = False
                break
        result.append(isprime)

    return result

is_prime([2,3,4,5,6,7,8,9])
```
