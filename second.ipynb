```python
# 컴퓨터와 대결하는 가위, 바위, 보 게임
# '그만'을 입력할때까지 계속
# '이겼습니다' '졌습니다' '비겼습니다'
import random

sel = ['가위' ,'바위' ,'보']
win = 0
lose = 0
draw = 0
while True:    
    com = random.randint(0,2)

    user = (input('가위바위보!'))
    if user == '그만':
        print('게임을 종료')
        break
    if user not in sel:
        print('제대로 입력하세요!!')
        continue
    if user == sel[com]:
        print('비겼어요')
        draw += 1
    elif(user == '가위' and sel[com] == '보') or (user == '바위' and sel[com] == '가위') or (user == '보' and sel[com] == '바위'):
        print('이겼어요')
        win += 1
    else:
    # elif(user == '가위' and sel[com] == '바위') or (user == '바위' and sel[com] == '보') or (user == '보' and sel[com] == '가위'):
        print('졌습니다')
        lose += 1
num = win + lose + draw
print(f'총 {num}전 {win}승 {draw}무 {lose}패')
```

```python
# 로또 번호 추첨기
# 1부터 45까지 6개의 중복되지 않는 번호를 추첨해서 append
# 오름차순으로 나열해서 출력 sort

import random
lotto = []

while True:
    if len(lotto) < 6 :
        com = random.randint(1,45)
        if com not in lotto:
            lotto.append(com)
    else:
        break
lotto.sort()
print(lotto)
```

```python
addr = {'박지민' : '010-1111-1234', '민윤기': '010-3333-1234', '김태형' : '010-5555-1234'}
while True:
    sel = input("메뉴 선택 1: 추가, 2 : 검색, 0: 종료")
    if sel == '1':
        addr.keys_name = input("이름을 입력하세요")
        add_num = input("전화번호를 입력하세요")
        addr = {'add_name' : 'add_num'}
    elif sel == '2':
        print(f'검색가능한 친구 {list(addr.keys())}')
        search_name = input("이름을 입력하세요")
        if search_name in addr:
            print(f'{search_name}의 연락처 {addr[search_name]}')
        else:
            print('등록되지 않은 친구입니다.')
    elif sel == '0':
        break
    else:
        print("메뉴 선택 오류입니다. 다시 선택하세요.")
print("프로그램을 종료합니다")
```

```python
morse = {
    '.-':'a','-...':'b','-.-.':'c','-..':'d','.':'e','..-.':'f',
    '--.':'g','....':'h','..':'i','.---':'j','-.-':'k','.-..':'l',
    '--':'m','-.':'n','---':'o','.--.':'p','--.-':'q','.-.':'r',
    '...':'s','-':'t','..-':'u','...-':'v','.--':'w','-..-':'x',
    '-.--':'y','--..':'z'
}

mor = input('모스구호를 입력하세요')
answer = ''

spl_mor = mor.split()

for i in spl_mor :
    # print(morse[i])
    answer += morse[i]

print(answer)
```
