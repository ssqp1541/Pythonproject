#### 조건문

```python
money = 1000
if money > 1500:
    print('커피를 마셔요')
    print('빵도 먹어요')
elif money > 500 :
    print('사탕')
else:
    print('물을 마셔요')
print('조건문 종료')
```

```python
# 조건문 활용 : 짝수 홀수 판단
# 2 -> 짝수 
# 3 -> 홀수

num = int(input("숫자를 입력하세요"))
if num % 2 == 0 :
    print("짝수입니다.")
else :
    print("홀수입니다.")
```

```python
# 성적처리 프로그램
# 90점 이상이면 'A'
# 80점 이상 90점 미만이면 'B'
# 70점 이상 80점 미만이면 'C'
# 60점 이상 70점 미만이면 'D'
# 아니면 'F'

my_score = int(input("성적을 입력하세요"))
if my_score >= 90 :
    print("A")
elif my_score >= 80 and my_score < 90 :
    print("B")
elif my_score >= 70 and my_score < 80 :
    print("C")
elif my_score >= 60 and my_score < 70 :
    print("D")
else :
    print("F")
```

```python
# 수학점수, 영어점수
# 수학 점수와 영어 점수가 둘다 90점 이상이면 ' 참잘햇어요'
# 수학 점수만 90점 이상이면 '좀 더 노력하세요 ' 
# 둘다 90점 미만이면 '형편 없네요'

math_score = int(input("수학 점수를 입력하세요"))
eng_score = int(input("영어 점수를 입력하세요"))

if math_score >= 90 and eng_score >= 90 :
    print("참 잘했어요.")
elif math_score >= 90 and eng_score < 90 : 
    print("영어 공부를 더 노력하세요")
elif eng_score >= 90 and math_score < 90 : 
    print("수학 공부를 더 노력하세요")
else :
    print("많이 노력하세요")
```

#### 반복문

```python
# for 변수 in 이터러블 :  * 이터러블 : 공이 여러개 들어있는 바구니
#     반복할 명령
```

```python
num = 1
for i in range(1,11,1):
    num *= i
print(num)
```

```python
# for 반복문과 range() 함수를 사용해서 입력받은 단의 구구단을 출력

for num in range(2,10,1) :
    print(f'{num}단')
    for i in range(1,10,1) :
        print(f'{num}*{i} = {num * i}')

```

```python
prompt = """"
1. ADD
2. Del
3. List
4. Quit

Enter number : """

user_num = 0
while user_num != 4:
    print(prompt)
    user_num = int(input("원하는 번호를 입력 : "))
    
```

```python
prompt = """"
1. ADD
2. Del
3. List
4. Quit

Enter number : """

user_num = 0
while True:
    print(prompt)
    user_num = int(input("원하는 번호를 입력 : "))
    if user_num == 4:
        break
```

```python
# 숫자 맞추기 게임

import random
num = random.randint(1,100)
num2 = int(input("숫자를 입력하세요"))

while True:
    if num2 == num :
        break
    elif num2 < num :
        print("업")
        num2 = int(input("숫자를 입력하세요"))
    else :
        print("다운")
        num2 = int(input("숫자를 입력하세요"))
        
    
print(f'정답입니다! 정답은 {num2}') 
```
