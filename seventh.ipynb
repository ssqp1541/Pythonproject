```python
import numpy as np
import pandas as pd


cities = pd.read_csv('data\\cities.csv')
constellations = pd.read_csv('data\\constellations.csv')
meteorshowers = pd.read_csv('data\\meteorshowers.csv')
moon = pd.read_csv('data\\moon_phases_2026.csv')
month = {'january':int(1), 'february':int(2), 'march':int(3), 'april':int(4), 'may':int(5), 'june':int(6), 'july':int(7),'august':int(8),'september':int(9),'october':int(10),'november':int(11),'december':int(12)}


meteorshowers['bestmonth'] = meteorshowers['bestmonth'].map(month)
meteorshowers['endmonth'] = meteorshowers['endmonth'].map(month)
meteorshowers['startmonth'] = meteorshowers['startmonth'].map(month)
constellations['bestmonth'] = constellations['bestmonth'].map(month)


# pd.to_datetime((meteorshowers['startmonth'])*100+(meteorshowers['startday'])+(2026*10000), format = '%Y%m%d')
meteorshowers['startdate'] = pd.to_datetime((meteorshowers['startmonth'])*100+(meteorshowers['startday'])+(2026*10000), format = '%Y%m%d')

# pd.to_datetime((2026*10000)+(meteorshowers['endmonth']*100)+meteorshowers['endday'], format = '%Y%m%d')
meteorshowers['enddate'] = pd.to_datetime((2026*10000)+(meteorshowers['endmonth']*100)+meteorshowers['endday'], format = '%Y%m%d')

# moon['moonphase'].unique()

phase = {'New Moon':0, 'First Quarter' : 0.5, 'Third Quarter':0.5, 'Full Moon': 1 }
moon['percentage'] = moon['moonphase'].map(phase)

constellations = constellations.drop(columns=['besttime'])
constellations

meteorshowers = meteorshowers.drop(columns=['startmonth','startday','endmonth','endday','preferredhemisphere'])

moon = moon.drop(columns=['moonphase','specialevent'])

moon
```

```python
moon['date'] = pd.to_datetime(moon['date'])
```

```python
moon['percentage'] = moon['percentage'].ffill().fillna(0.5)

moon
```

```python
def star(city):
    meteor_shower_string = ''
    if city not in cities['city'].values:
        meteor_shower_string = f'아쉽지만 {city}에는 별자리 서비스가 제공되지 않습니다!.'
        return meteor_shower_string
    
    
    latitude = cities[cities['city'] == city]['latitude'][0]
    constellations_list = constellations[(constellations['latitudestart'] >= latitude) & (constellations['latitudeend'] <= latitude)]['constellation'].to_list()
    for i in constellations_list:
        meteorshower = meteorshowers.loc[meteorshowers['radiant'] == i,'name'].iloc[0]
        starttime = meteorshowers.loc[meteorshowers['radiant'] == i,'startdate'].iloc[0]
        endtime = meteorshowers.loc[meteorshowers['radiant'] == i,'enddate'].iloc[0]
        moon_list = moon[(moon['date'] >= starttime) & (moon['date'] <= endtime)]
        best_moon_date = moon_list.loc[moon_list['percentage'].idxmin(),'date']
        best_moon_date_str = best_moon_date.to_pydatetime().strftime('%Y년 %m월 %d일')
        meteor_shower_string += f'{city} 에서 {best_moon_date_str}에 {meteorshower} 를 보실수 있습니다.\n'
    return meteor_shower_string

print(star('Abu Dhabi'))
```
