```python
import requests
from bs4 import BeautifulSoup
# 1. 대상 URL 설정
url = "https://news.naver.com/section/105" # IT/과학 뉴스
# 2. HTTP GET 요청 (User-Agent 설정은 필수 - 봇 차단 방지)
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}
response = requests.get(url, headers=headers)
# 3. HTML 파싱
soup = BeautifulSoup(response.text, "html.parser")
# 4. 데이터 추출 (CSS Selector 활용)
# 네이버 뉴스 구조에 따라 class명은 변경될 수 있음
news_list = soup.select(".sa_text_strong")
print(f"--- 오늘의 IT 뉴스 헤드라인 ({len(news_list)}건) ---")
for idx, news in enumerate(news_list[:10], 1):
    print(f"{idx}. {news.get_text().strip()}")



```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html,'html.parser')

articles = soup.select(".block1")

data = []

for article in articles:
    title = article.select_one(".articleSubject").text.strip()
    summary = article.select_one(".articleSummary").contents[0].strip()
    press = article.select_one(".press").text.strip()
    wdate = article.select_one(".wdate").text
    link = article.select_one(".thumb > a").attrs['href']


    office_id = link.split("=")[2].split("&")[0]
    article_id = link.split("=")[1].split("&")[0]

    link = f"https://n.news.naver.com/mnews/article/{office_id}/{article_id}"
    data.append({"title":title, "summary":summary, "press":press, "wdate":wdate, "link":link})

from datetime import datetime
date = datetime.now().strftime('%y-%m-%d')

import pandas as pd
df = pd.DataFrame(data)
file_name = f'data/네이버_오늘의_뉴스_{date}'
df.to_csv(f'{file_name}.csv')
```

```python
import requests
import xmltodict


url = "https://apis.data.go.kr/1613000/RTMSDataSvcAptTrade/getRTMSDataSvcAptTrade"
params = {
    'LAWD_CD' : '44133',
    'DEAL_YMD': '201810',
    } 

response = requests.get(url,params=params)

response_data = xmltodict.parse(response.content)


items = response_data['response']['body']['items']['item']
item = items[0]
data = []

for item in items :
    data.append({'아파트명':item['aptNm'], '법정동':item['umdNm'],'거래금액':item['dealAmount'],
    '전용면적':item['excluUseAr'],'층':item['floor'],'건축년도':item['buildYear']})

import pandas as pd
df = pd.DataFrame(data)
df
file_name = f'data/천안_아파트_실거래가_{params["DEAL_YMD"]}'
df.to_csv(f'{file_name}.csv')
# # # aptNm
```
